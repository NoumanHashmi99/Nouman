```python
def avg(file,col_index):
    lst = []
    for i in file:
        value = i[col_index]
        value = float(value)
        lst.append(value)
        a = sum(lst)/len(lst)
        return a
```

```python
avg(dataset,4)
```

```python
def freq(file,col_index):
    dic = {}
    for i in file:
        value = i[col_index]
        if value not in dic:
            dic[value] = 1
        else:
            dic[value] +=1
    return dic
```

```python
freq(dataset,2)
```
